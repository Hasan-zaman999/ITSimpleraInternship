Load the Dataset and Display Basic Information

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset directly from the UCI Machine Learning Repository
# Note: The Online Retail II dataset is an Excel file containing multiple sheets.
# We will load the first sheet (Year 2009-2010) for this analysis.
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00502/online_retail_II.xlsx"
df = pd.read_excel(url, sheet_name="Year 2009-2010")

# 2. Display basic information
print("--- Dataset Shape ---")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n--- Dataset Columns ---")
print(df.columns.tolist())

print("\n--- Data Types and Non-Null Counts ---")
print(df.info())

print("\n--- First 5 Rows of the Data ---")
df.head()
```

Identify and Report Missing Values and Duplicate Rows

```python
# 1. Identify and report missing values
print("--- Missing Values Per Column ---")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

print("\n--- Percentage of Missing Values ---")
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage[missing_percentage > 0])

# 2. Identify and report duplicate rows
duplicate_count = df.duplicated().sum()
print("\n--- Duplicate Rows ---")
print(f"Total number of duplicate rows found: {duplicate_count}")
```

Analyze the Top 10 Best-Selling Products by Quantity and Revenue

```python
# Create a Revenue column (Quantity multiplied by Price)
df['Revenue'] = df['Quantity'] * df['Price']

# 1. Top 10 Best-Selling Products by Quantity
top_quantity = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

# 2. Top 10 Best-Selling Products by Revenue
top_revenue = df.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)

print("--- Top 10 Products by Quantity ---")
print(top_quantity)

print("\n--- Top 10 Products by Revenue ---")
print(top_revenue)

# Plotting the results
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.barplot(x=top_quantity.values, y=top_quantity.index, palette="viridis")
plt.title("Top 10 Products by Total Quantity Sold", fontsize=12, fontweight='bold')
plt.xlabel("Total Quantity")
plt.ylabel("Product Description")

plt.subplot(1, 2, 2)
sns.barplot(x=top_revenue.values, y=top_revenue.index, palette="magma")
plt.title("Top 10 Products by Total Revenue Generated", fontsize=12, fontweight='bold')
plt.xlabel("Total Revenue (£)")
plt.ylabel("")

plt.tight_layout()
plt.show()
```

Analyze Sales Performance by Country

```python
# Group by country to calculate total revenue and total quantity ordered
country_analysis = df.groupby('Country').agg({
    'Quantity': 'sum',
    'Revenue': 'sum'
}).sort_values(by='Revenue', ascending=False)

print("--- Sales Performance by Country (Top 10) ---")
print(country_analysis.head(10))

# Plotting top 10 countries by revenue
plt.figure(figsize=(12, 6))
sns.barplot(x=country_analysis['Revenue'].head(10).values, y=country_analysis['Revenue'].head(10).index, palette="Blues_r")
plt.title("Top 10 Countries by Total Revenue", fontsize=14, fontweight='bold')
plt.xlabel("Total Revenue (£)")
plt.ylabel("Country")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

Plot Revenue Over Time (Monthly Trend)

```python
# Extract Year-Month from InvoiceDate for monthly grouping
df['InvoiceYearMonth'] = df['InvoiceDate'].dt.to_period('M')

# Group by month and calculate sum of revenue
monthly_revenue = df.groupby('InvoiceYearMonth')['Revenue'].sum()

# Convert index back to timestamp for seamless plotting with matplotlib
monthly_revenue.index = monthly_revenue.index.to_timestamp()

# Plot monthly trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o', color='b', linestyle='-', linewidth=2)
plt.title("Monthly Revenue Trend (2009 - 2011)", fontsize=14, fontweight='bold')
plt.xlabel("Timeline (Months)")
plt.ylabel("Total Revenue (£)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```

Create a Correlation Heatmap

```python
# Select only numeric variables for correlation analysis
numeric_cols = df[['Quantity', 'Price', 'Revenue']]

# Calculate the correlation matrix
correlation_matrix = numeric_cols.corr()

print("--- Correlation Matrix ---")
print(correlation_matrix)

# Plot the Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Numeric Variables", fontsize=14, fontweight='bold')
plt.show()
```

Detect and Visualize Outliers Using Box Plots

```python
# Plot boxplots to visualize extreme value distributions (outliers)
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, y='Quantity', color='skyblue')
plt.title("Outlier Detection in Quantity", fontsize=12, fontweight='bold')
plt.ylabel("Quantity Values")

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='Price', color='salmon')
plt.title("Outlier Detection in Price", fontsize=12, fontweight='bold')
plt.ylabel("Price Values (£)")

plt.tight_layout()
plt.show()
```

Business Insights Documented From the Data

## Final Deliverable: 5 Crucial Business Insights Derived From the Analysis

### 1. Massive Domestic Market Concentration (The UK Hub)
* **Observation:** The United Kingdom alone accounts for **8.19 million £** out of the total revenue, while the next largest country market (EIRE) accounts for only **352,242 £**. The UK also commands over **4.4 million units** in quantity sold.
* **Business Implication:** The business is heavily dependent on the UK economic landscape. While it represents a highly successful core market, it presents a risk of over-reliance.
* **Actionable Recommendation:** Management should maintain its dominant position in the UK through loyalty rewards while actively targeting expansion or localized marketing strategies in high-potential European markets like EIRE, the Netherlands, and Germany to diversify revenue streams.

---

### 2. High Guest-Checkout Volume & Relationship Gaps
* **Observation:** The missing value analysis shows that **20.54% of rows lack a `Customer ID`**, whereas only **0.56% lack a product `Description`**.
* **Business Implication:** Over one-fifth of all transactions are untracked at the user level, likely due to guest checkouts, missing registrations, or systemic omissions. This creates a blind spot for customer lifetime value (CLV) metrics, personalized retention campaigns, and recommendation engine training.
* **Actionable Recommendation:** Implement post-purchase incentives (e.g., *"Create an account to track your order and save 5% on your next purchase"*) to convert guest buyers into registered profiles without adding friction to the checkout pipeline.

---

### 3. Divergence Between Volume Drivers and Premium Revenue Generators
* **Observation:** Product volume does not directly equate to cash generation. For example, **"WORLD WAR 2 GLIDERS ASSTD DESIGNS"** is the 2nd highest item by volume (**54,698 units**) but does not make the top 10 revenue list. Conversely, **"REGENCY CAKESTAND 3 TIER"** yields the absolute highest revenue (**163,051.46 £**) despite not being the top volume driver.
* **Business Implication:** The inventory consists of two distinct product profiles: high-turnover/low-margin convenience items that pull traffic to the site, and premium/high-margin destination products that capture actual dollar value.
* **Actionable Recommendation:** Cross-merchandise these segments. Bundle or recommend high-revenue destination products (like the Regency Cakestand) alongside high-frequency volume drivers to optimize average order value (AOV).

---

### 4. Severe Transaction Anomalies (Returns & Cancellations)
* **Observation:** The outlier boxplots show massive data spreads extending deep into negative territories, with quantities dropping past **-5,000** and price variables showing substantial negative fluctuations down to **-50,000**.
* **Business Implication:** These extreme negative values indicate a high volume of cancelled orders, damaged freight write-offs, or commercial returns. Furthermore, the correlation between `Price` and `Revenue` ($r = 0.45$) is drastically diluted by these structural anomalies.
* **Actionable Recommendation:** Audit the transaction logs to isolate cancellation codes from actual sales. Before training any machine learning models, implement a data-cleaning pipeline that handles return entries explicitly so they do not skew customer purchasing patterns.

---

### 5. Highly Seasonal, Q4-Dependent Sales Trajectory
* **Observation:** The monthly revenue trend plot reveals a steady baseline hovering between **0.5M £ and 0.8M £** for most of the year, followed by an exponential surge starting in September (**~0.85M £**), peaking dramatically in November at over **1.4M £**, and plunging sharply in December.
* **Business Implication:** The business is highly cyclical and relies heavily on holiday shopping (Q4 peak). A poor performance in October and November could destabilize the company's annual profit targets.
* **Actionable Recommendation:** Optimize supply chain logistics and inventory levels by August to prevent stockouts during the holiday rush. Additionally, design mid-year marketing events (e.g., *"Christmas in July"* promotions) to smooth out revenue dips experienced during summer lulls (such as July's drop to **~0.58M £**).
